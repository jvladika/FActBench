```python
import glob
extension = 'csv'
results_dir = "results/"
runs = glob.glob('{}/*.{}'.format(results_dir, extension))
len(runs)
```

```python
import pandas as pd
columns = ['id', 'model', 'ds','task', 'path']
df = pd.DataFrame(columns = columns)
for run in runs:
    #df = pd.read_csv(run) 
    idd = run.split("/")[1].split('_')[0]
    model = run.split("/")[1].split('_')[2]
    ds =  run.split("/")[1].split('_')[4]
    task =  run.split("/")[1].split('_')[6]
    df.loc[len(df)] = [idd,model,ds,task,run]

df['model'].value_counts()
```
